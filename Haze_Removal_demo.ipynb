```
import os
import cv2
from google.colab.patches import cv2_imshow
path1="/content/drive/My Drive/SIH/SIH Videos/HazyImages/Input"
path2="/content/drive/My Drive/SIH/SIH Videos/HazyImages/Output"
counter=1
while counter<=10:
  
  for input in os.listdir(path1):
    if input == "image"+str(counter)+".jpg":
      print("Hazy Image")
      inp=cv2.imread(path1+'/'+input)
      cv2_imshow(inp)

  for output in os.listdir(path2):
    if output == "out"+str(counter)+".jpg":
      print("Dehazed Image")
      oup=cv2.imread(path2+'/'+output)
      cv2_imshow(oup)

  counter+=1

  
```
